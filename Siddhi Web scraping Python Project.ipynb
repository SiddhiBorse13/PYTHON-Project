```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/crypto/?count=100&offset=0'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify
```

```python
table=soup.find('table')
```

```python
table=table.find('tbody')
```

```python
table
```

```python
rows=table.find_all('tr')
rows
```

```python
rows=table.find_all('tr',style="")
rows
```

```python
column_info=[]
for i in rows:
    cols=i.find_all('td')
    #print cols
    country_info=[c.text for c in cols]
    #print country_info
    column_info.append(country_info)
print(column_info)
```

```python
Symbol=[]
Name=[]
Price=[]
Change=[]
percentage_Change=[]
Market_Cap=[]
Volume_in_Currency_UTC=[]
Volume_in_Currency_24Hr=[]
Total_Volume_All_Currencies=[]
Circulating_Supply=[]
for i in column_info:
    Symbol.append(i[0])
    Name.append(i[1])
    Price.append(i[2])
    Change.append(i[3])
    percentage_Change.append(i[4])
    Market_Cap.append(i[5])
    Volume_in_Currency_UTC.append(i[6])
    Volume_in_Currency_24Hr.append(i[7])
    Total_Volume_All_Currencies.append(i[8])
    Circulating_Supply.append(i[9])
```

```python
Data1={'Symbol':Symbol,'Name':Name,'Price':Price,'Change':Change,'percentage_Change':percentage_Change,'Market_Cap':Market_Cap,'Volume_in_Currency_UTC':Volume_in_Currency_UTC,'Volume_in_Currency_24Hr':Volume_in_Currency_24Hr,'Total_Volume_All_Currencies':Total_Volume_All_Currencies,'Circulating_Supply':Circulating_Supply}
```

```python
crypto_data=pd.DataFrame(Data1)
crypto_data
```

```python
crypto_data.to_csv('crypto.csv')
```
